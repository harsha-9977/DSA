```python
# Linked List - Finding the Critical Task in a Project Workflow (get From middle)
class Solution:
  class Node:
    def __init__(self,data):
      self.data = data
      self.next = None

  class LinkedList:
    def __init__(self):
      self.head = None

    def insertAtEnd(self,data):
      newNode = Solution.Node(data)

      if self.head is None:
        self.head = newNode
      else:
        curr = self.head
        while curr.next is not None:
          curr = curr.next
        curr.next = newNode

    def printlist(self):
      curr = self.head
      while curr is not None:
        print(curr.data , end=" ")
        curr = curr.next

    def getFromMiddle(self):
      slow, fast = self.head, self.head

      while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

      result = []
      while slow is not None:
        result.append(slow.data)
        slow = slow.next
      return result

if __name__ == "__main__":
  n = int(input())

  li = list(map(int, input().split()))

  if len(li) != n:
    print("length is not same")
  else:
    lists = Solution.LinkedList()

    for i in li:
      lists.insertAtEnd(i)

    output = lists.getFromMiddle()
    print(*output)
```

```python
# Linked List - Reorganizing Job Priorities in a Task Manager (swapping kth from start and end also)
class Solution:
  class Node:
    def __init__(self,data):
      self.data = data
      self.next = None

  class LinkedList:
    def __init__(self):
      self.head = None

    def insertAtEnd(self,data):
      newNode = Solution.Node(data)

      if self.head is None:
        self.head = newNode
      else:
        curr = self.head
        while curr.next is not None:
          curr = curr.next
        curr.next = newNode

    def printlist(self):
      curr = self.head
      while curr is not None:
        print(curr.data , end=" ")
        curr = curr.next

    def size(self):
      curr = self.head

      count = 0
      while curr is not None:
        count += 1
        curr = curr.next
      return count

    def swapkth(self,k):
      n = self.size()

      prevX, currX = None, self.head
      for i in range(1,k):
        prevX = currX
        currX = currX.next

      prevY, currY = None, self.head
      for i in range(1,n-k+1):
        prevY = currY
        currY = currY.next

      if prevX is not None:
        prevX.next = currY
      else:
        self.head = currY

      if prevY is not None:
        prevY.next = currX
      else:
        self.head = currX

      # still i have to swap next of currX and currY

      currX.next, currY.next = currY.next, currX.next

if __name__ == "__main__":
  n = int(input())

  li = list(map(int, input().split()))

  if len(li) != n:
    print("Oops!")
  else:
    lists = Solution.LinkedList()

    for i in li:
      lists.insertAtEnd(i)

    lists.swapkth(int(input()))
    lists.printlist()
```
