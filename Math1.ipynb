```python
n = 10
a = 5
d = 3

sn = n * ((2 * a) + ((n-1) * d)) // 2
print(sn)

s1 = 0
for i in range(1,n+1):
  s1 += (a+(i-1)*d)
  print(s1)
```

```python
# Functions

def f(x):
  return x**3 + 1

def g(x):
  return np.sqrt(x+2)

print("(g o f)(2) = ", g(f(2)))
print("(f o g)(2) = ", f(g(2)))
print("(g o g)(2) = ", g(g(2)))
```

```python
# y1 = 3^x, y2 = 3^(x-1) + 2

x = np.linspace(-1,2,5)
y1 = 3**x
y2 = 3**(x-1) + 2

graph1, = plt.plot(x,y1)
graph2, = plt.plot(x,y2)
plt.legend(handles=(graph1,graph2), labels=(r'$y=3^x$', r'$y=3^{x-1}+2$'))
plt.show()
```

```python
# y1 = 2^-x, y2 = 2^-x-2

x = np.linspace(-1,2,5)
y1 = 2**-x
y2 = -2**(-x-2)

graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$2^{-x}$', r'$2^{-x-2}$'))
plt.show()
```

```python
# Logarithms
# y= log2(x+1) - 1

x1 = np.linspace(0.1,4,401)
y1 = np.log2(x1)
x2 = np.linspace(-0.9,4,501)
y2 = np.log2(x2+1)-1

graph1, = plt.plot(x1, y1)
graph2, = plt.plot(x2, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$log_2(x)$', r'$log_2(x+1)-1$'))
plt.show()
```
