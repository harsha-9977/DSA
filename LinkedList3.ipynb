```python
# Linked List - Swapping Tasks in Pairs in a Scheduler
class Solution:
  class Node:
    def __init__(self,data):
      self.data = data
      self.next = None

  class LinkedList:
    def __init__(self):
      self.head = None

    def insertAtEnd(self,data):
      newNode = Solution.Node(data);

      if self.head is None:
        self.head = newNode
      else:
        curr = self.head
        while curr.next is not None:
          curr = curr.next
        curr.next = newNode

    def printlist(self):
      curr = self.head
      while curr is not None:
        print(curr.data , end = " ")
        curr = curr.next

    def swapPairs(self):
      dummy = Solution.Node(0)
      dummy.next = self.head
      prev = dummy

      while prev.next is not None and prev.next.next is not None:
        first = prev.next
        second = prev.next.next

        first.next = second.next
        second.next = first
        prev.next = second

        prev = first

      self.head = dummy.next

if __name__ == "__main__":
  n = int(input())
  li = list(map(int, input().split()))

  if len(li) != n:
    print("length not matching")
  else:
    lists = Solution.LinkedList()

    for i in li:
      lists.insertAtEnd(i)

    lists.swapPairs()
    lists.printlist()
```

```python
# Linked List - Merging Patient Queues in a Hospital
class Solution:
  class Node:
    def __init__(self,data):
      self.data = data
      self.next = None

  class LinkedList:
    def __init__(self):
      self.head = None

    def insertAtEnd(self,data):
      newNode = Solution.Node(data)

      if self.head is None:
        self.head = newNode
      else:
        curr = self.head
        while curr.next is not None:
          curr = curr.next
        curr.next = newNode

    def printlist(self):
      curr = self.head
      while curr is not None:
        print(curr.data , end = " ")
        curr = curr.next

    def mergeList(self,l1,l2):
      mergedList = Solution.LinkedList()
      dummy = Solution.Node(0)
      tail = dummy

      ptr1 = l1.head
      ptr2 = l2.head

      while ptr1 is not None and ptr2 is not None:
        if ptr1.data <= ptr2.data:
          tail.next = ptr1
          ptr1 = ptr1.next
        else:
          tail.next = ptr2
          ptr2 = ptr2.next
        tail = tail.next

      if ptr1 is not None:
        tail.next = ptr1
      elif ptr2 is not None:
        tail.next = ptr2

      mergedList.head = dummy.next
      return mergedList

if __name__ == "__main__":
  n = int(input())
  li1 = list(map(int, input().split()))

  if len(li1) != n:
    print("Oops")
  else:
    list1 = Solution.LinkedList()
    for i in li1:
      list1.insertAtEnd(i)

  m = int(input())
  li2 = list(map(int, input().split()))

  if len(li2) != m:
    print("Oops")
  else:
    list2 = Solution.LinkedList()
    for i in li2:
      list2.insertAtEnd(i)

  mergedList = Solution.LinkedList().mergeList(list1,list2)
  mergedList.printlist()
```
