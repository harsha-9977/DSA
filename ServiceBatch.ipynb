Pangram Check for Automated Book Scanner System
Imagine you are working on a project for a library that is developing an automated book scanning system. The system needs to determine if a given sentence from a book contains all the letters of the English alphabet at least once. If a sentence does contain every letter from 'a' to 'z', it is called a pangram. You have been tasked with implementing a solution that uses recursion to check for pangrams.

You are required to implement a program that uses a recursive function to check if a given string is a pangram. A string is considered a pangram if it contains every letter of the English alphabet at least once, regardless of case.



Input Format

The input consists of a single line containing the string to be checked.
The string can contain uppercase or lowercase letters, spaces, and punctuation marks.


Output Format

Print "Pangram" if the string contains all the letters of the alphabet.
Print "Not Pangram" if it does not.


Example 1

Sample Input 1

The quick brown fox jumps over the lazy dog



Sample Output 1

Pangram



Explanation

The given sentence contains every letter of the English alphabet at least once.



Example 2

Sample Input 2

Hello World



Sample Output 2

Not Pangram



Explanation

The given sentence does not contain every letter of the English alphabet.

```python
def isPan(s):
  alpha = "abcdefghijklmnopqrstuvwxyz"
  for i in alpha:
    if i not in s:
      return False
  return True

n = input().strip().lower()

if(isPan(n)):
  print("Pangram")
else:
  print("Not Pangram")
```

You are given a requirement to print seat numbers in a pyramid-like pattern. The seat numbers should follow a specific sequence, where:

The first row has 1 seat.
The second row has 2 seats, but they are displayed in descending order.
The third row has 3 seats in ascending order.
The fourth row has 4 seats in descending order.
The fifth row has 5 seats in ascending order, and so on.
Write a program to generate this seat numbering pattern.



Input Format

An integer n which indicates the number of rows for the seat numbering pattern.


Output Format

Print the seat numbers in the specified pattern.


Constraints

1 ≤ n ≤ 10 (The number of rows will not exceed 10)


Sample Input

5



Sample Output

1

3 * 2

4 * 5 * 6

10 * 9 * 8* 7

11 * 12 * 13 * 14 * 15

```python
n = int(input())
start = 1
for i in range(1,n+1):
  st = ""
  if i%2 == 0:
    for j in range(i):
      st = str(start) + "*" + st
      start += 1
    st = st[:-1]
  else:
    for j in range(i):
      st = st + "*" + str(start)
      start += 1
    st = st[1:]
  print(st)
```

Finding Kaprekar Numbers for Math Enthusiasts
Jaffer is preparing for his upcoming Math Olympiad and has been fascinated by unique mathematical concepts. His Math teacher, Meena, has introduced him to Kaprekar numbers. A Kaprekar number is defined as follows:

Consider an n-digit number k. If you square k, split the result into two parts such that:
The right part has n digits (or less if necessary), and the left part has the remaining digits.
If the sum of these two parts equals k, then k is a Kaprekar number.
Jaffer needs your help to determine whether a given number is a Kaprekar number.



Input Format

A single integer that represents the number k.


Constraints

There are no specific constraints.


Output Format

Print "Kaprekar Number" if the number is a Kaprekar number.
Otherwise, print "Not a Kaprekar Number".


Example 1

Sample Input 1

45



Sample Output 1

Kaprekar Number



Explanation

The number 45 squared is 2025. When split into two parts, 20 and 25, the sum is 20 + 25 = 45, which equals the original number. Thus, 45 is a Kaprekar number.



Example 2

Sample Input 2

23



Sample Output 2

Not a Kaprekar Number



Explanation

The number 23 squared is 529. When split into parts (5 and 29), the sum is 5 + 29 = 34, which is not equal to the original number 23. Thus, 23 is not a Kaprekar number.

```python
def isK(n):
  o = n
  n = str(n**2)
  m = len(n) // 2
  s1 = n[:m]
  s2 = n[m:]

  return (int(s1) + int(s2)) == o

n = int(input().strip())

if isK(n):
  print("Kaprekar Number")
else:
  print("Not a Kaprekar Number")
```

100

1

10

Interest Calculation:

A single integer n, representing the number of rows in the triangle.


Output Format

Print the palindrome triangle pattern.

   1

Each row contains a sequence of numbers that form a palindrome.

The fourth row has 4 seats in descending order.

Write a program to generate this seat numbering pattern.



Input Format

An integer n which indicates the number of rows for the seat numbering pattern.


Output Format

Print the seat numbers in the specified pattern.


Constraints

1 ≤ n ≤ 10 (The number of rows will not exceed 10)


Sample Input

5



Sample Output

1

3 * 2

4 * 5 * 6

10 * 9 * 8 * 7

11 * 12 * 13 * 14 * 15

```python
n = int(input())
val = 1
for i in range(1,n+1):
  st = ""
  if i%2 == 0:
    for j in range(i):
      st = str(val) + "*" + st
      val += 1
    print(st[:-1])
  else:
    for j in range(i):
      st = st + "*" + str(val)
      val += 1
    print(st[1:])
  print()
```
